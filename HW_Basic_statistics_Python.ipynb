## Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
# Переводим все значения датафрейма в числовые. NaN оставляем, не учитывается при расчетах и не искажает статистику
df = df.apply(pd.to_numeric)
df.sort_values('pain').head(200)
```

```python
# Выводим все базовые статистики по всем столбцам, по строчке 'count' понимаем, где больше пробелов в данных
df.describe()
```

####  Смотрим выбросы для столбцов с большим отклонением (std) -  Age, Pulse, Respiratory rate

```python
# Выводим данные по возрасту, которые лежат внутри первого и третьего процентиля
q1_age = df['Age'].quantile(0.25)
q3_age = df['Age'].quantile(0.75)
iqr_age = q3_age - q1_age
lower_bound_age = q1_age - (1.5 * iqr_age) 
upper_bound_age = q3_age + (1.5 * iqr_age)
remove_outliers_age = df[df['Age'].between(lower_bound_age, upper_bound_age, inclusive = True)].sort_values('Age')
remove_outliers_age.tail()
```

```python
# Выбросы по возрасту = результат ошибки, так как возможны только значения 1 = Adult horse и 2 = Young horse
# Возможно, 9 должны были быть 2, так как в таблице выше есть только Age = 1. Нужно проверить и после этого или заменить или убрать из датафрейма, если лошади данного возраста не были целью исследования.
pd.concat([df, remove_outliers_age]).drop_duplicates(keep = False)
```

```python
# 24 cтроки из 300 -- это 8% от всей выборки
pd.concat([df, remove_outliers_age]).drop_duplicates(keep = False).info()
```

```python
# Выводим данные по пульсу, которые лежат внутри первого и третьего процентиля
q1_pulse = df['pulse'].quantile(0.25)
q3_pulse = df['pulse'].quantile(0.75)
iqr_pulse = q3_pulse - q1_pulse
lower_bound_pulse = q1_pulse - (1.5 * iqr_pulse) 
upper_bound_pulse = q3_pulse + (1.5 * iqr_pulse)
remove_outliers_pulse = df[df['pulse'].between(lower_bound_pulse, upper_bound_pulse, inclusive = True)].sort_values('pulse')
remove_outliers_pulse.head()
```

## Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# Количество пропусков по столбцам
print(df.isna().sum())
```

```python
# Количество пропусков по всему датафрейму
# Важно! Предполагаем все пропуски случайными
print(df.isna().sum().sum())
```

```python
# Age - если в случае с возрастом понимаем, что 9 - это ошибка и должно быть 2, то производим замену
df = df.replace({'Age':{9:2}}).sort_values('Age')
df.tail()
```

```python
import numpy as np
```

```python
# surgery, outcome  - значение отсутствует для одной и той же строчки, то есть нет ни данных об операции, ни о результате, удаляем
df = df[np.logical_not(np.isnan(df.surgery))]
df = df.reset_index(drop = True)
df.sort_values('surgery').tail()
```

```python
# rectal temperature - решила заменить NaN на среднюю температуру, так как стандартное отклонение небольшое
print(df.groupby(['outcome', 'surgery'])['rectal temperature'].mean())
df['rectal temperature'].fillna(df['rectal temperature'].mean(), inplace = True)
df['rectal temperature'].fillna(df['rectal temperature'].mean()).isna().sum()
```

```python
# temperature of extremities - заменяю медианой, объяснить не могу, возможно, можно и среднее поставить
print(df.groupby(['outcome', 'surgery'])['temperature of extremities'].median())
df['temperature of extremities'].fillna(df['temperature of extremities'].median(), inplace = True)
df['temperature of extremities'].fillna(df['temperature of extremities'].median()).isna().sum()
```

```python
# pain - заменила бы на среднее или медиану, так как отклонение (std) небольшое
print(df.groupby(['surgery', 'temperature of extremities'])['pain'].median())
df['pain'].fillna(df['pain'].median(), inplace = True)
df['pain'].fillna(df['pain'].median()).isna().sum()
```

```python
# respiratory rate - заменила на медиану, так как много выбросов
print(df.groupby(['Age'])['respiratory rate'].median())
df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace = True)
df['respiratory rate'].fillna(df['respiratory rate'].median()).isna().sum()
```

```python
# pulse - коррелирует с temperature of extremities и pain, решила заполнить на основании категорий
print(df.groupby([ 'Age', 'temperature of extremities','pain'])['pulse'].median())
df['pulse'].fillna(df.groupby(['Age', 'temperature of extremities', 'pain'])['pulse'].transform('median'), inplace=True)
```

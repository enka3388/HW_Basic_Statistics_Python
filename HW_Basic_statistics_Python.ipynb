## Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.
https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.names

```python
import pandas as pd
```

```python
# загружаем данные со столбцами без названия, используя header = None
df = pd.read_csv('horse_data.csv', header = None)
df.head()
```

```python
# столбцы, которые нам нужны
columns_list = [0, 1, 3, 4, 5, 6, 10 ,22]
```

```python
# оставляем только 8 столбцов, которые нужны по условию задания, удалив остальные
for col in df.columns:

    if col not in columns_list:

        del df[col]
df.head()
```

```python
# присваиваем названия столбцам
df.columns = ['surgery', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']
df.head()
```

## Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
# заменяем знаки ? на пустые значения
def change_item(item):
    if item == '?':
        item = ''
    return item
df =  df.applymap(change_item)
df.head()
```

```python
# Переводим все значения датафрейма в числовые, в пустых ячейках получим NaN, который не учитывается при расчетах и не искажает статистику
df = df.apply(pd.to_numeric)
df.sort_values('pulse').head(200)
```

```python
# Выводим все базовые статистики по всем столбцам, по строчке 'count' понимаем, где больше пробелов в данных
df.describe()
```

####  Смотрим выбросы для столбцов с большим отклонением (std) -  Age, Pulse, Respiratory rate

```python
# Выводим данные по возрасту, которые лежат внутри первого и третьего процентиля
q1_age = df['Age'].quantile(0.25)
q3_age = df['Age'].quantile(0.75)
iqr_age = q3_age - q1_age
lower_bound_age = q1_age - (1.5 * iqr_age) 
upper_bound_age = q3_age + (1.5 * iqr_age)
remove_outliers_age = df[df['Age'].between(lower_bound_age, upper_bound_age, inclusive = True)].sort_values('Age')
remove_outliers_age.tail()
```

```python
df.info()
```

```python
# Выбросы по возрасту = результат ошибки, так как возможны только значения 1 = Adult horse и 2 = Young horse
# Возможно, 9 должны были быть 2, так как в таблице выше есть только Age = 1. Нужно проверить и после этого или заменить или убрать из датафрейма, если лошади данного возраста не были целью исследования.
pd.concat([df, remove_outliers_age]).drop_duplicates(keep = False)
```

```python
# 24 cтроки из 300 -- это 8% от всей выборки
pd.concat([df, remove_outliers_age]).drop_duplicates(keep = False).info()
```

```python
# Выводим данные по пульсу, которые лежат внутри первого и третьего процентиля
q1_pulse = df['pulse'].quantile(0.25)
q3_pulse = df['pulse'].quantile(0.75)
iqr_pulse = q3_pulse - q1_pulse
lower_bound_pulse = q1_pulse - (1.5 * iqr_pulse) 
upper_bound_pulse = q3_pulse + (1.5 * iqr_pulse)
remove_outliers_pulse = df[df['pulse'].between(lower_bound_pulse, upper_bound_pulse, inclusive = True)].sort_values('pulse')
remove_outliers_pulse.head()
```

```python
# Выбросы по пульсу = по факту только 5 лошадей с выбросом из-за высокого пульса, остальное -- это отсутствующие данные. Можно оставить NaN, так как не искажают вычисления.
pd.concat([df, remove_outliers_pulse]).drop_duplicates(keep = False).sort_values('pulse', ascending = False)
```

```python
# Выводим данные по частоте дыхания, которые лежат внутри первого и третьего процентиля
q1_resp = df['respiratory rate'].quantile(0.25)
q3_resp = df['respiratory rate'].quantile(0.75)
iqr_resp = q3_resp - q1_resp
lower_bound_resp = q1_resp - (1.5 * iqr_resp) 
upper_bound_resp = q3_resp + (1.5 * iqr_resp)
remove_outliers_resp = df[df['respiratory rate'].between(lower_bound_resp, upper_bound_resp, inclusive = True)].sort_values('respiratory rate')
remove_outliers_resp.head()
```

```python
# Выбросы по частоте дыхания.
pd.concat([df, remove_outliers_resp]).drop_duplicates(keep = False).sort_values('respiratory rate', ascending = False)
```
